## Pie chart 3

**Algo más complejo**

Características:

   * Se indican los colores
   * le digo que empiece a plotear desde un ángulo de 90º
   * radio del pie chart 1.0

```python
# plt.pie?
```

```python
y = ['Big Data', 'Algoritmos', 'Machine Learning', 'Visualización de Datos'] 
x = [4, 2, 4, 1] 

plt.title("Horas semanales de clase")
plt.pie(x, labels = y,
        colors = ["r", "y", "g", "b"],
        startangle = 90,
        radius = 1)

plt.show()
```

## Pie chart 4

**autopct le indica que escriba el número dentro (%)**

```python
y = ['Big Data', 'Algoritmos', 'Machine Learning', 'Visualización de Datos'] 
x = [4, 2, 4, 1] 

plt.title("Horas semanales de clase")
plt.pie(x, labels = y,
        colors = ["r", "y", "g", "b"],
        startangle = 90,
        radius = 1.1,
        autopct="%1.2f%%")

plt.show()
```

# Varios plots

```python
import numpy as np
plt.figure(figsize=(12,8))   
# lo metes todo en una figura para que no se te quede pequeño..

x = np.arange(-10, 10, 0.01) 
# simplemente le das el rango (-10,10) de 0.01 en 0.01

# simplemente le dices los subplots 
# (número de filas, número de columnas, número de grafico)
# GRÁFICO 1
plt.subplot(2,2,1) 
y1=x
plt.plot(x, y1)
plt.title("y=x")

# GRÁFICO 2
plt.subplot(2,2,2) 
y2=x**2
plt.plot(x, y2)
plt.title("y es x al cuadrado")

# GRÁFICO 3
plt.subplot(2,2,3) 
y3=x**3
plt.plot(x, y3)
plt.title("y es x al cubo")

# GRÁFICO 
plt.subplot(2,2,4) 
y4=x**4
plt.plot(x, y4)
plt.title("y es x a la cuarta")

plt.show()
```
